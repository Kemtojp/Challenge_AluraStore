# Análisis de Tiendas - Alura Store

Este informe tiene como objetivo ayudar al Sr. Juan a tomar una decisión estratégica respecto a cuál de sus cuatro tiendas debería vender, basándose en el análisis de rendimiento de cada una.

Se trabajó con los conjuntos de datos correspondientes a cada tienda, utilizando herramientas de análisis como Pandas, Matplotlib y Seaborn para obtener métricas clave. A continuación se detallan los hallazgos.



### Importación de datos



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



En esta sección se calculan los ingresos generados por cada tienda, sumando el total de precios de los productos vendidos.

Este indicador nos permite identificar qué tienda presenta mayor rentabilidad en términos de ventas.

```python
ingreso_t1 = round(tienda["Precio"].sum())
ingreso_t2 = round(tienda2["Precio"].sum())
ingreso_t3 = round(tienda3["Precio"].sum())
ingreso_t4 = round(tienda4["Precio"].sum())
```

```python
print("Ingreso por tienda: ")
print("Tienda 1:", ingreso_t1)
print("Tienda 2:", ingreso_t2)
print("Tienda 3:", ingreso_t3)
print("Tienda 4:", ingreso_t4)
```

# 2. Ventas por categoría

Aquí agrupamos las ventas por categoría de producto para cada tienda.

Este análisis también ayuda a entender el comportamiento del consumidor y el foco comercial de cada tienda.


```python
ventas_categoria_t1 = tienda["Categoría del Producto"].value_counts()
ventas_categoria_t2 = tienda2["Categoría del Producto"].value_counts()
ventas_categoria_t3 = tienda3["Categoría del Producto"].value_counts()
ventas_categoria_t4 = tienda4["Categoría del Producto"].value_counts()
```

```python
print("Productos por categoria de cada tienda: ")
print("Tienda 1:\n", ventas_categoria_t1)
print("Tienda 2:\n", ventas_categoria_t2)
print("Tienda 3:\n", ventas_categoria_t3)
print("Tienda 4:\n", ventas_categoria_t4)
```

# 3. Calificación promedio de la tienda


Esta sección refleja la satisfacción del cliente a través del promedio de calificaciones otorgadas en cada tienda.

Una mayor calificación indica una mejor experiencia de compra.


```python
calificacion_t1 = round(tienda["Calificación"].mean(), 2)
calificacion_t2 = round(tienda2["Calificación"].mean(), 2)
calificacion_t3 = round(tienda3["Calificación"].mean(), 2)
calificacion_t4 = round(tienda4["Calificación"].mean(), 2)
```

```python
print("Calificacion promedio por tienda:")
print("Tienda 1: ", calificacion_t1)
print("Tienda 2: ", calificacion_t2)
print("Tienda 3: ", calificacion_t3)
print("Tienda 4: ", calificacion_t4)
```

# 4. Productos más y menos vendidos

Permite identificar las categorías más populares y menos solicitadas.

```python
print("Categorias más populares por tienda: ")
print("Tienda 1: ", ventas_categoria_t1.idxmax(), "-", ventas_categoria_t1.max())
print("Tienda 2: ", ventas_categoria_t2.idxmax(), "-", ventas_categoria_t2.max())
print("Tienda 3: ", ventas_categoria_t3.idxmax(), "-", ventas_categoria_t3.max())
print("Tienda 4: ", ventas_categoria_t4.idxmax(), "-", ventas_categoria_t4.max())
```

```python
print("Categorias menos populares por tienda: ")
print("Tienda 1: ", ventas_categoria_t1.idxmin(), "-", ventas_categoria_t1.min())
print("Tienda 2: ", ventas_categoria_t2.idxmin(), "-", ventas_categoria_t2.min())
print("Tienda 3: ", ventas_categoria_t3.idxmin(), "-", ventas_categoria_t3.min())
print("Tienda 4: ", ventas_categoria_t4.idxmin(), "-", ventas_categoria_t4.min())
```

# 5. Envío promedio por tienda

Se calcula el valor promedio del costo de envío en cada tienda.

Este análisis ayuda a comprender la eficiencia logística de cada punto de venta.


```python
costo_envio_t1 = round(tienda["Costo de envío"].mean())
costo_envio_t2 = round(tienda2["Costo de envío"].mean())
costo_envio_t3 = round(tienda3["Costo de envío"].mean())
costo_envio_t4 = round(tienda4["Costo de envío"].mean())
```

```python
print("Costo de envio promedio por tienda:")
print("Tienda 1: ", costo_envio_t1)
print("Tienda 2: ", costo_envio_t2)
print("Tienda 3: ", costo_envio_t3)
print("Tienda 4: ", costo_envio_t4)
```

#Visualizacion

```python
tienda["tienda"] = "Tienda 1"
tienda2["tienda"] = "Tienda 2"
tienda3["tienda"] = "Tienda 3"
tienda4["tienda"] = "Tienda 4"

df = pd.concat([tienda, tienda2, tienda3, tienda4], ignore_index=True)
```

```python
# Calcular los ingresos totales por tienda
ingresos_por_tienda = df.groupby("tienda")["Precio"].sum()

# Graficar
plt.figure(figsize=(8,5))
ingresos_por_tienda.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Ingresos Totales por Tienda")
plt.ylabel("Ingreso Total ($)")
plt.xlabel("Tienda")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

La **Tienda 1** presenta el ingreso más alto, mientras que la **Tienda 4** se encuentra en el último lugar.


```python
import seaborn as sns

# Contar productos por categoría y tienda
categoria_por_tienda = df.groupby(["tienda", "Categoría del Producto"]).size().reset_index(name="Cantidad")

# Crear el gráfico
plt.figure(figsize=(12,6))
sns.barplot(data=categoria_por_tienda, x="Categoría del Producto", y="Cantidad", hue="tienda")
plt.title("Cantidad de Productos Vendidos por Categoría y Tienda")
plt.xticks(rotation=45)
plt.ylabel("Cantidad de Productos")
plt.xlabel("Categoría")
plt.legend(title="Tienda")
plt.tight_layout()
plt.show()
```

Se observa una clara tendencia: **"Muebles" y "Electrónicos"** son las categorías más populares en todas las tiendas, siendo las que más unidades concentran. Por el contrario, **"Artículos para el hogar"** o **"Instrumentos musicales"** suelen estar entre las menos vendidas.


```python
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
calificaciones = [3.98, 4.04, 4.05, 4.00]

plt.figure(figsize=(8, 5))
plt.plot(tiendas, calificaciones, marker='o', linestyle='-', color='mediumseagreen')
plt.title('Calificación Promedio por Tienda')
plt.ylabel('Calificación Promedio')
plt.xlabel('Tienda')
plt.ylim(3.8, 4.1)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

La **Tienda 3** lidera en satisfacción con una calificación promedio de **4.05**, seguida muy de cerca por Tienda 2. La **Tienda 1** tiene la más baja (**3.98**), aunque ninguna tienda cae por debajo del estándar esperado.


```python
costos_envio = [26019, 25216, 24806, 23459]

plt.figure(figsize=(7, 7))
plt.pie(costos_envio, labels=tiendas, autopct='%1.1f%%', startangle=140,
        colors=['#ff9999','#ffc000','#8fd9b6','#d395d0'])
plt.title('Distribución del Costo de Envío Promedio por Tienda')
plt.axis('equal')
plt.tight_layout()
plt.show()

```

La **Tienda 4** destaca como la más eficiente en costos de envío, con un promedio más bajo que el resto. En contraste, la **Tienda 1** tiene el costo más alto, lo cual puede afectar la rentabilidad general.


# Conclusión

Tras analizar múltiples indicadores clave, se concluye lo siguiente:

- **Ingresos:** Tienda 4 es la que menos genera.
- **Calificación del cliente:** Tienda 3 se destaca levemente por sobre las demás.
- **Costo de envío:** Tienda 4 es la más eficiente en términos logísticos.
- **Popularidad de productos:** La tendencia se mantiene similar en todas las tiendas, con "Muebles" y "Electrónicos" liderando las ventas.
